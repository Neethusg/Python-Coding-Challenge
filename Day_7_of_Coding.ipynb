```python
import matplotlib.pyplot as plt

# Regional sales data
region_sales = {
    'North': 300,
    'South': 450,
    'East': 250,
    'West': 500
}

regions = list(region_sales.keys())
sales = list(region_sales.values())

# Find top-performing region
top_region = regions[sales.index(max(sales))]

# Set color scheme: highlight the top region
colors = ['skyblue' if region != top_region else 'orange' for region in regions]

plt.figure(figsize=(10, 6))
plt.barh(regions, sales, color=colors, edgecolor='black')
plt.xlabel('Units Sold')
plt.title('Sales by Region')
plt.gca().invert_yaxis()  # So highest region appears on top

# Add labels to bars
for i, value in enumerate(sales):
    plt.text(value + 10, i, str(value), va='center')

plt.tight_layout()
plt.show()
```
