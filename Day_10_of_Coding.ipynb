```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Raw scores dictionary
scores = {
    "Alice": {"Math": 88, "Science": 92, "History": 75, "English": 85},
    "Bob": {"Math": 76, "Science": 85, "History": 80, "English": 70},
    "Charlie": {"Math": 90, "Science": 87, "History": 85, "English": 88},
    "Diana": {"Math": 65, "Science": 78, "History": 70, "English": 72},
    "Ethan": {"Math": 82, "Science": 89, "History": 90, "English": 85}
}

# Convert to DataFrame
df = pd.DataFrame(scores).T  # Transpose to get students as rows

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="coolwarm", fmt='d', linewidths=0.5, cbar=True)

# Formatting
plt.title("Student Test Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Student")
plt.tight_layout()

plt.show()
```
