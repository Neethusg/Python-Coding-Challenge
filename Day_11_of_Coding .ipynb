```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data for sales in three regions over 12 months
months = np.arange(1, 13)
sales_north = np.random.randint(15000, 25000, 12)
sales_south = np.random.randint(12000, 22000, 12)
sales_west = np.random.randint(10000, 20000, 12)

# Generate meshgrid for the surface plot
X, Z = np.meshgrid(months, np.array([1, 2, 3]))  # X is months, Z is regions (1=North, 2=South, 3=West)

# Create corresponding Y values for each region
Y_north = sales_north
Y_south = sales_south
Y_west = sales_west

# Stack these values into a single array for the surface plot
Y = np.vstack([Y_north, Y_south, Y_west])

# Create figure and 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Z, Y, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_xlabel('Month')
ax.set_ylabel('Region')
ax.set_zlabel('Sales ($)')
ax.set_title('3D Surface Plot of Sales by Region')

# Add a color bar
mappable = ax.plot_surface(X, Z, Y, cmap='viridis', edgecolor='none')
fig.colorbar(mappable, shrink=0.5, aspect=5)

# Show plot
plt.tight_layout()
plt.show()
```
