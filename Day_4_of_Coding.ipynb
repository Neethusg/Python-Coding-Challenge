Challenge: Visualize Sales Data

```python
sales_data = {
    'Laptops': 150,
    'Tablets': 90,
    'Smartphones': 200,
    'Headphones': 120,
    'Monitors': 60
}
```

```python
import matplotlib.pyplot as plt

def visualize_sales(sales_data):
    products = list(sales_data.keys())
    sales = list(sales_data.values())

    # Bar Chart
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.bar(products, sales, color='skyblue', edgecolor='black')
    plt.title('Total Sales per Product')
    plt.xlabel('Product')
    plt.ylabel('Units Sold')
    plt.xticks(rotation=45)

    # Pie Chart
    plt.subplot(1, 2, 2)
    plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=140)
    plt.title('Sales Distribution (%)')

    plt.tight_layout()
    plt.show()

sales_data = {
    'Laptops': 150,
    'Tablets': 90,
    'Smartphones': 200,
    'Headphones': 120,
    'Monitors': 60
}
visualize_sales(sales_data)
```
