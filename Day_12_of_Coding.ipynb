```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulating a large dataset for 1000 products over 12 months
num_products = 1000
num_months = 12

# Generate random sales data (values between 1000 and 50000)
sales_data = np.random.randint(1000, 50000, (num_products, num_months))

# Create product names (Product_1, Product_2, ..., Product_1000)
product_names = [f"Product_{i+1}" for i in range(num_products)]

# Create a DataFrame for easier handling and analysis
df_sales = pd.DataFrame(sales_data, columns=[f"Month_{i+1}" for i in range(num_months)], index=product_names)

# 1. **Histogram** of total sales across all products
total_sales = df_sales.sum(axis=1)
plt.figure(figsize=(10, 6))
plt.hist(total_sales, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Total Sales Across All Products")
plt.xlabel("Total Sales ($)")
plt.ylabel("Number of Products")
plt.tight_layout()
plt.show()

# 2. **Line Chart** showing sales trends for 5 random products
random_products = np.random.choice(df_sales.index, size=5, replace=False)
plt.figure(figsize=(12, 6))
for product in random_products:
    plt.plot(df_sales.columns, df_sales.loc[product], label=product)
plt.title("Sales Trends for 5 Random Products")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. **Box Plot** to show distribution of sales across all months
plt.figure(figsize=(12, 6))
plt.boxplot([df_sales[f"Month_{i+1}"] for i in range(num_months)], labels=[f"Month_{i+1}" for i in range(num_months)])
plt.title("Sales Distribution Across Months")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.tight_layout()
plt.show()
```
