```python
import matplotlib.pyplot as plt

# Customer segmentation data
customer_data = {
    'New': 50,
    'Returning': 120,
    'VIP': 30,
    'Churned': 20
}

labels = list(customer_data.keys())
values = list(customer_data.values())

# Highlight the 'VIP' segment by exploding it
explode = [0.05 if label == 'VIP' else 0 for label in labels]
colors = ['#8ecae6', '#219ebc', '#ffb703', '#fb8500']  # Custom color palette

plt.figure(figsize=(8, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode, colors=colors)
plt.title('Customer Segmentation')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle

plt.tight_layout()
plt.show()
```
