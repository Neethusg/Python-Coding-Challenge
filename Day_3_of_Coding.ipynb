Python Challenge: Plot a Custom Sine Wave

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sine_wave(freq, amp, duration):
    sample_rate = 1000
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y = amp * np.sin(2 * np.pi * freq * t)

    plt.figure(figsize=(10, 4))
    plt.plot(t, y, label=f'{freq}Hz Sine Wave')
    plt.title('Sine Wave')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_sine_wave(freq=5, amp=2, duration=2)
```
