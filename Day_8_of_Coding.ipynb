```python
import matplotlib.pyplot as plt

traffic_data = {
    "Week 1": {"Desktop": 3400, "Mobile": 5600, "Tablet": 1200},
    "Week 2": {"Desktop": 3100, "Mobile": 5900, "Tablet": 1000},
    "Week 3": {"Desktop": 4000, "Mobile": 6200, "Tablet": 1400},
    "Week 4": {"Desktop": 3700, "Mobile": 5800, "Tablet": 1100},
}

# Sum traffic for each device type
device_totals = {}
for week in traffic_data.values():
    for device, visits in week.items():
        device_totals[device] = device_totals.get(device, 0) + visits

# Sort by total visits
sorted_devices = sorted(device_totals.items(), key=lambda x: x[1], reverse=True)
devices = [item[0] for item in sorted_devices]
totals = [item[1] for item in sorted_devices]

# Colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, Orange, Green

# Plot
plt.figure(figsize=(8, 5))
bars = plt.barh(devices, totals, color=colors)

# Add data labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 100, bar.get_y() + bar.get_height() / 2,
             f'{width}', va='center')

# Formatting
plt.title("Total Website Traffic by Device Type")
plt.xlabel("Total Visits")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

# Show plot
plt.show()
```
