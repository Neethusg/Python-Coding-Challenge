```python
import matplotlib.pyplot as plt

# Sample revenue data
revenue_data = {
    'Jan': 12000,
    'Feb': 15000,
    'Mar': 17000,
    'Apr': 13000,
    'May': 16000,
    'Jun': 18000
}

months = list(revenue_data.keys())
revenue = list(revenue_data.values())

# Find highest and lowest
max_rev = max(revenue)
min_rev = min(revenue)
max_month = months[revenue.index(max_rev)]
min_month = months[revenue.index(min_rev)]

plt.figure(figsize=(10, 6))
plt.plot(months, revenue, marker='o', linestyle='-', color='teal')
plt.title('Monthly Revenue Trend (First Half)')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True)

# Annotate highest and lowest points
plt.annotate(f'Highest: {max_month}\n${max_rev}',
             xy=(max_month, max_rev),
             xytext=(max_month, max_rev + 2000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             ha='center')

plt.annotate(f'Lowest: {min_month}\n${min_rev}',
             xy=(min_month, min_rev),
             xytext=(min_month, min_rev - 3000),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center')

plt.tight_layout()
plt.show()
```
