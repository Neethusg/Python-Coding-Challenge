```python
import matplotlib.pyplot as plt

# Expense data
expenses = {
    "Rent": 1200,
    "Groceries": 450,
    "Utilities": 200,
    "Transport": 150,
    "Entertainment": 300,
    "Dining Out": 250,
    "Healthcare": 100,
    "Other": 50
}

# Extract data
categories = list(expenses.keys())
amounts = list(expenses.values())

# Find index of largest category
max_index = amounts.index(max(amounts))
explode = [0.1 if i == max_index else 0 for i in range(len(amounts))]

# Colors
colors = plt.cm.Paired(range(len(categories)))

# Plot
plt.figure(figsize=(7, 7))
plt.pie(amounts,
        labels=categories,
        autopct='%1.1f%%',
        startangle=140,
        explode=explode,
        colors=colors,
        shadow=True)

# Title and format
plt.title("Monthly Expenses Breakdown")
plt.axis('equal')  # Ensure pie is circular

# Show plot
plt.show()
```
